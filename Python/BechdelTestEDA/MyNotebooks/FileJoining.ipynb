```python
tables1 = pd.read_html("https://www.boxofficemojo.com/alltime/world/?pagenum=1&p=.htm")
testing1 = tables1[1]

tables2 = pd.read_html("https://www.boxofficemojo.com/alltime/world/?pagenum=2&p=.htm")
testing2 = tables2[1]

tables3 = pd.read_html("https://www.boxofficemojo.com/alltime/world/?pagenum=3&p=.htm")
testing3 = tables3[1]

tables4 = pd.read_html("https://www.boxofficemojo.com/alltime/world/?pagenum=4&p=.htm")
testing4 = tables4[1]

tables5 = pd.read_html("https://www.boxofficemojo.com/alltime/world/?pagenum=5&p=.htm")
testing5 = tables5[1]

tables6 = pd.read_html("https://www.boxofficemojo.com/alltime/world/?pagenum=6&p=.htm")
testing6 = tables6[1]

tables7 = pd.read_html("https://www.boxofficemojo.com/alltime/world/?pagenum=7&p=.htm")
testing7 = tables7[1]

tables8 = pd.read_html("https://www.boxofficemojo.com/alltime/world/?pagenum=8&p=.htm")
testing8 = tables8[1]

merged_df = pd.concat([testing1, testing2, testing3, testing4, testing5, testing6, testing7, testing8])
```

```python
movies_mojo_list = merged_df.values.tolist()
print(movies_mojo_list[0])
```

```python
headers = movies_mojo_list[0]
print(headers)

for i in range(8):
    movies_mojo_list.remove(headers)

movies_df = pd.DataFrame(movies_mojo_list)
movies_df.columns = ['Rank', 'primaryTitle', 'Studio', 'Worldwide', 'Domestic', "Domestic_Perc", 'Overseas', "Overseas_Perc", 'Year']
movies_df.head()
```

```python
#remove ^ from Year
def release_year(year):
    if len(str(year))>4:
        return str(year)[:4]
    else:
        return str(year)

movies_df['Year'] = movies_df.Year.apply(release_year)
movies_df.head()
```

```python
#create new merge key TitleYear for movies_df
movies_df['TitleYear'] = movies_df.primaryTitle + movies_df.Year
movies_df.head()
```

```python
#create new merge key TitleYear for bechdel_titlebasics_ratings
bechdel_titlebasics_ratings['TitleYear'] = bechdel_titlebasics_ratings.primaryTitle + bechdel_titlebasics_ratings.startYear
bechdel_titlebasics_ratings.head()
```

```python
bechdel_titlebasics_ratings_mojo = bechdel_titlebasics_ratings.merge(movies_df, on = 'TitleYear')
```
