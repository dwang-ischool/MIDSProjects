```python
%matplotlib inline
import sys
print(sys.version)
import numpy as np
print(np.__version__)
import pandas as pd
print(pd.__version__)
import matplotlib.pyplot as plt
import csv
```

```python
#dataframe of top 50 bankable actors
actors_df = pd.read_csv('../Data/Top50BankableActors_nconst.csv', index_col = 0)
```

```python
#dataframe of top 50 bankable actors on IMDB
actors_movies_df = pd.read_csv('../Data/T50BA_all_movies.csv', index_col = 0)
```

```python
#all imdb movies with bechdel ratings
titles_df = pd.read_csv('../Data/bechdel_titlebasics.csv', index_col = 0)
```

```python
actors_df.head(10)
```

```python
#make rank a column:
actors_df['BoxOffice_Rank'] = actors_df.index
```

```python
actors_df.head(10)
```

```python
actors_movies_df.head(10)
```

```python
#checks and filtering
actors_movies_df.groupby('titletype').sum() 
```

titletype requires no filtering since only movie and tvMovies are included, which aligns with our other analysis

```python
actors_movies_df.groupby('is_adult').sum() 
```

no additional filtering for adult films is required, since no adult movies in this dataframe

```python
#actors_movies_df is the product of 2 merged IMDB tables.  First remove all of the titles that we do not need:
actors_movies_df.drop(['ordering', 'char_played', 'titletype', 'originaltitle','is_adult','start_yr','end_yr','runtimeminutes', 'genres', 'job', 'category'], axis = 1, inplace = True)
```

```python
actors_movies_df.head(10)
```

```python
titles_df.head(10)
```

```python
#some filtering
titles_df = titles_df[(titles_df.titleType == 'movie') | (titles_df.titleType == 'tvMovie')]
titles_df = titles_df[titles_df.isAdult == 0]
```

```python
#removing unnecessary columns:
titles_df.drop(['title', 'release_yr', 'titleType', 'originalTitle', 'bechdel_id','isAdult','startYear','endYear','runtimeMinutes', 'genres'], axis = 1, inplace = True)
titles_df.head(10)
```

```python
#rename columns to prepare for joining:

titles_df.columns = ['imdb_id', 'bechdel_rating', 'TitleBechdel']
actors_movies_df.columns = ['nconst', 'imdb_id', 'TitleIMDB']

titles_df.set_index('imdb_id', inplace = True)
actors_movies_df.set_index('imdb_id', inplace = True)
```

```python
titles_df.head()
```

```python
#join titles_df and actors_movies_df to get bechdel ratings of the actors movies.  Please note, some actors movies will not have bechdel ratings

actors_movies_ratings_df = actors_movies_df.join(titles_df)
```

```python
actors_movies_ratings_df.head(20)
```

Merge appears to have worked correctly.  We shoud have a table of all titles that an actor has starred in.  For movies that have bechdel ratings, the merge should have pulled those ratings.  For all other actors' movies that do not have bechdel ratings, TitleBechdel and Bechdel_Rating should return NaN.  The TitleIMDB should match TitleBechdel if a movie has a bechdel rating.

```python
#merge actors_movies_ratings_df and actors_df
actors_movies_ratings_df.set_index('nconst', inplace = True)
actors_df.set_index('nconst', inplace = True)
```

```python
actors_movies_ratings_df.head(10)
```

```python
#final merge to pull actors name and rank to list of movies with bechdel ratings:

t50actorsmovies_df = actors_movies_ratings_df.join(actors_df)
t50actorsmovies_df.head(10)
```

```python
#check the data
t50actorsmovies_df.groupby('primaryname').sum()
```

```python
#going to remove some columns and add new columns for analysis:
t50actorsmovies_df.drop(['TitleIMDB','TitleBechdel','Num_1_Picture', 'Gross_Millions'], axis = 1, inplace = True)
t50actorsmovies_df['Rank_Name'] = '('+t50actorsmovies_df['BoxOffice_Rank'].astype(str)+', '+t50actorsmovies_df['primaryname'].astype(str)+')'
t50actorsmovies_df.head(10)
```

```python
#add counter for grouping
t50actorsmovies_df['Movies_Count'] = 1
t50actorsmovies_df = t50actorsmovies_df.sort_values('BoxOffice_Rank', ascending = False)
t50actorsmovies_df['bechdel_rating'] = t50actorsmovies_df['bechdel_rating'].fillna('None')
t50actorsmovies_df.head(10)
```

```python
actors_avg_gross = t50actorsmovies_df[['Rank_Name', 'Average_Millions']].drop_duplicates()
actors_avg_gross.head(50)
```

```python
actors_avg_gross.describe()
```

```python
#in order to plot this table, need to turn the average millions into numbers

avg_gross = lambda x: float(x[1:])

actors_avg_gross['Average_Gross'] = actors_avg_gross.Average_Millions.apply(avg_gross)
actors_avg_gross.set_index('Rank_Name', inplace=True)
```

```python
plt.rc('xtick', labelsize=12) 
plt.rc('ytick', labelsize=12) 
actors_gross_bar = actors_avg_gross.plot(kind='barh', stacked=True, colormap='tab20', figsize = (15,20))
actors_gross_bar.set_xlabel('Average Gross (Millions $/Movie)', size=20)
actors_gross_bar.set_ylabel('Rank, Actor/Actress', size=20)
actors_gross_bar.set_title('Box Office Mojo\'s Top 50 Grossing Actors\' Gross per Film', size=22)

actors_gross_bar
```

```python
actors_table_df = t50actorsmovies_df.groupby(['Rank_Name', 'bechdel_rating'])['Movies_Count'].sum().unstack('bechdel_rating')
actors_table_df.head(50)
```

```python
#sort in order:
actors_table_df['pair'] = actors_table_df.index
actors_table_df['order'] = actors_table_df['pair'].astype(str).apply(lambda s: int(s[1: s.find(',')]))
actors_table_df = actors_table_df.sort_values('order', ascending = False)
actors_table_df.drop(['pair', 'order'], axis = 1, inplace = True)
actors_table_df.head(50)
```

```python
plt.rc('xtick', labelsize=12) 
plt.rc('ytick', labelsize=12) 
actors_bechdel_bar = actors_table_df.plot(kind='barh', stacked=True, colormap='RdYlGn', figsize = (15,20))
actors_bechdel_bar.set_xlabel('Number of Films', size=20)
actors_bechdel_bar.set_ylabel('Rank, Actor/Actress', size=20)
actors_bechdel_bar.set_title('Box Office Mojo\'s Top 50 Grossing Actors\' Bechdel Test Movie Ratings', size=23)

actors_bechdel_bar
```

```python
#remove None column for heat map:
actors_bechdel_table = actors_table_df.drop('None', axis = 1)
actors_bechdel_table.columns = ['0', '1', '2', '3']
actors_bechdel_table = actors_bechdel_table.sort_values('3', ascending = False)
actors_bechdel_table.head(10)
```

```python
import seaborn as sns
plt.figure(figsize=(15,20))
bechdel_actors_heat = sns.heatmap(actors_bechdel_table, cmap = 'BuGn', cbar_kws={'label': 'Number of Movies'}, annot=True)
bechdel_actors_heat.set_xlabel('Bechdel Test Rating', size=20)
bechdel_actors_heat.set_ylabel('Rank, Actors/Actresses', size=20)
bechdel_actors_heat.set_title('Bechdel Test Ratings for Top 50 Grossing Actors', size=25)
bechdel_actors_heat.figure.axes[-1].yaxis.label.set_size(20)
```

```python
fig1 = actors_gross_bar.get_figure()
fig1.savefig('actors_gross_bar.png', bbox_inches='tight')

fig2 = actors_bechdel_bar.get_figure()
fig2.savefig('actors_bechdel_bar.png', bbox_inches='tight')

fig3 = bechdel_actors_heat.get_figure()
fig3.savefig('bechdel_actors_heat.png', bbox_inches='tight')
```

```python
actors_percent_none = actors_table_df.copy().fillna(0.)
actors_percent_none.columns = ['0','1','2','3','None']
actors_percent_none['Percent None'] = actors_percent_none['None'] / (actors_percent_none['0']+actors_percent_none['1']+actors_percent_none['2']+actors_percent_none['3']+actors_percent_none['None'])*100
actors_percent_none['Percent None'].describe()
```

The following was intended to see if there is any correlation between percentage of Bechdel Test Passing films and Average Gross.  
Unfortunately, there is none, so the following will not be included in the report

```python
actors_percent_pass = actors_table_df.copy().fillna(0.)
actors_percent_pass.columns = ['0','1','2','3','None']
actors_percent_pass['Percent Pass'] = actors_percent_pass['3'] / (actors_percent_pass['0']+actors_percent_pass['1']+actors_percent_pass['2']+actors_percent_pass['3'])*100
actors_percent_pass['pair'] = actors_percent_none.index
actors_percent_pass['order'] = actors_percent_pass['pair'].astype(str).apply(lambda s: int(s[1: s.find(',')]))
actors_percent_pass = actors_percent_pass.sort_values('order')
actors_percent_pass.drop(['pair', 'order'], axis = 1, inplace = True)
actors_percent_pass.head(50)
```

```python
actors_percent_pass_avg_gross = actors_avg_gross.join(actors_percent_pass)
```

```python
actors_percent_pass_avg_gross.head()
```

```python
actors_percent_pass_avg_gross.drop(['0', '1', '2', '3', 'None', 'Average_Millions'], axis = 1, inplace = True)
```

```python
plt.rc('xtick', labelsize=12) 
plt.rc('ytick', labelsize=12) 
gross_pass_bar = actors_percent_pass_avg_gross.plot(kind='barh', colormap='RdYlGn', figsize = (15,20))
gross_pass_bar.set_xlabel('Number of Films', size=20)
gross_pass_bar.set_ylabel('Rank, Actor/Actress', size=20)
gross_pass_bar.set_title('Box Office Mojo\'s Top 50 Grossing Actors\' Bechdel Test Movie Ratings', size=23)

gross_pass_bar
```
